```python
def catagorized_inches(x):
    if x<13.3 :
        return "Small"
    elif 13.3 <= x <15.6 :
        return "Medium"
    else :
        return "Large"

Data["catagorized_inches"]=Data.Inches.apply(catagorized_inches)
    
```
